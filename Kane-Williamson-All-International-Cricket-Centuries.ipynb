```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
data = pd.read_csv(r"E:\Projects Datas\Sports\Kane-Williamson-All-International-Cricket-Centuries.csv")
data.head()
```

```python
data.info()
```

```python
#we have already has an index
data.drop('No.',axis=1,inplace = True)
```

```python
data['Not_out'] = data['Score'].str.contains(r'\*')
```

```python
# Converting Score into int
data['Score'] = data['Score'].str.replace('*', '',regex=True)
data['Score'] = data['Score'].astype(int)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data
```

```python
data.info()
```

```python
from pandas_profiling import ProfileReport
Report = ProfileReport(data)
Report.to_file(output_file='Kane.html')
```

```python
# No of International-Cricket-Centuries
data.shape[0]
```

```python
# Highest Score
data[data.Score == data.Score.max()]
```

```python
#No of Not_outs
plt.pie(data.Not_out.value_counts(),autopct='%1.1f%%',labels=data.Not_out.value_counts().index)
plt.show()
print(data.Not_out.value_counts())
```

```python
#avg score above 100
data.Score.mean()
```

```python
#Ground
plt.pie(data.Ground.value_counts(),autopct='%1.1f%%',labels=data.Ground.value_counts().index)
plt.show()
print(data.Ground.value_counts())
```

```python
#Most Centuries vs Country
plt.figure(figsize=(5,6))
plt.barh(y = data.Against.value_counts().index,width=data.Against.value_counts())
plt.show()
```

```python
match_result = data.groupby(['Against', 'Result']).size().unstack()
match_result.plot(kind='bar',stacked = True,figsize =(15,5))
plt.ylabel('no_of_Centrys',color='blue',font='Comic Sans MS', size=10)
plt.show()
```

```python
# year vs centures
plt.figure(figsize=(20,5))
sns.lineplot(data=data,x='Date',y='Score',color = 'blue',marker='*',markersize=10)
plt.show()
```

```python
year = data.groupby(data.Date.dt.year).Score.count()
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x=year.index,y=year,color='darkblue')
plt.show()
```
