```python
from sklearn.preprocessing import StandardScaler
df_scaled = StandardScaler().fit_transform(df)
```

```python
df_scaled.shape
```

```python
df1 = df_scaled.T
```

```python
df1
```

```python
cov_mat = np.cov(df1)
cov_mat.shape
```

```python
eig_values, eig_vecs = np.linalg.eig(cov_mat)
```

```python
eig_values
```

```python
eig_vecs
```

```python
tot = sum(eig_values)
var_exp = [(i/tot)*100 for i in eig_values]
var_exp
```

```python
cum_var_exp = np.cumsum(var_exp)
cum_var_exp
```

```python
plt.figure(figsize=(10, 5))
plt.bar(range(len(var_exp)), var_exp, label='individual explained variance', color = 'g')
plt.step(range(len(cum_var_exp)), cum_var_exp,label='cumulative explained variance')
plt.ylabel('Explained variance ')
plt.xlabel('Principal components')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans 
```

```python
pca = PCA(n_components=10)# almost 90%
```

```python
df_scaled = pca.fit_transform(df_scaled)
```

```python
clusters = list(range(2,9))
ssd = []
for num_clusters in clusters:
    model_clus = KMeans(n_clusters = num_clusters, max_iter=150 )
    model_clus.fit(df_scaled)
    ssd.append(model_clus.inertia_)
```

```python
plt.plot(clusters,ssd)
plt.ylabel("SSD")
```

from here we get 3 Clusters

```python
from sklearn.metrics import silhouette_score
import numpy as np

def find_optimal_clusters(data, max_clusters):
    silhouette_scores = []
    
    for n_clusters in range(2, max_clusters+1):
        kmeans = KMeans(n_clusters=n_clusters)
        labels = kmeans.fit_predict(data)
        silhouette_avg = silhouette_score(data, labels)
        silhouette_scores.append(silhouette_avg)
    
    optimal_clusters = np.argmax(silhouette_scores) + 2  # Adding 2 to align with range(2, max_clusters+1)
    return optimal_clusters

# Example usage
max_clusters = 9

optimal_clusters = find_optimal_clusters(df_scaled, max_clusters)
print("Optimal number of clusters:", optimal_clusters)
```
